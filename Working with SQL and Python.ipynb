```python
import mysql.connector as connection
import pandas as pandas
import csv
try:
    mydb = connection.connect(host = "127.0.0.1", user = "root", passwd = "Qwerty123", use_pure = True)
    print(mydb.is_connected())
    query = "create database db1;"
    cursor = mydb.cursor()
    cursor.execute(query)
    print("database created")
    mydb.close()
    
    mydb = connection.connect(host = "127.0.0.1", database = "db1", user = "root", passwd = "Qwerty123", 
                              use_pure = True)
    query = "create table churnmodelling(CustomerId int(10), Surname varchar(20), CreditScore int(10),"\
             "Geography varchar(10), Gender varchar(10), Age int(10), Tenure int(10))"
    cursor = mydb.cursor()
    cursor.execute(query)
    print("table created")
    
    with open('churnmodelling.data', "r") as f:
        next(f)
        churnmodelling_data = csv.reader(f, delimiter="\n")
        for line in enumerate(churnmodelling_data):
            for list_ in (line[1]):
                cursor.execute("insert into churnmodelling({values})".format(values=(list_)))
        print("values inserted") 
        mydb.commit()
        cursor.close()
        mydb.close()
except Exception as e:
    mydb.close()
    print(str(e))
    
```

```python
def sql_insert(path, db, tab):
    with open(str(path), 'r') as data:
        file = csv.reader(data, delimiter = ',')
        cur = mydb.cursor()
        next(file)
        for data in file:
            final_data = tuple([f'{row}' for row in data]) 
            query = f"insert into {db}.{tab} values{final_data};"
#             print(query)
            cur.execute(query)
        mydb.commit()
        print('value inserted into sql')
```
